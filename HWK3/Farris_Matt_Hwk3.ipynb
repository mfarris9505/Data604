### Part A: 

```python
import numpy as np
import pandas as pd

def dice_roll (number):
    """
    Fair roll of 2 dices: Inputs are the number of rolls, 
    and the output is the number on each dice and the total sum. 
    """
    dice_1 = []
    dice_2 = [] 
    sum_dice =[]
    for i in range(1,number+1):    
        x = np.random.randint(1,6)
        y = np.random.randint(1,6)
        z = x+y
        dice_1.append(x)
        dice_2.append(y)
        sum_dice.append(z)


    result = pd.DataFrame({"Dice 1":dice_1,"Dice 2":dice_2,"Roll":sum_dice})
    result.index += 1
    return result
```

```python
result_50 = dice_roll(50)
result_500 = dice_roll(500)
```

A quick peek at the top ten rolls shows us the following: 

```python
result_50.head(10)
result_500.head(10)
```

As you can see from the above, the distribution at 500 shows more of a uniformed distribution centering around 7 rolls. This makes sense as the most common roll given any 2 dice will be a total of 7.

```python
def weighted_dice (number, prob):
    """
    Unfair roll of 2 dices: Inputs are the number of rolls, 
    and the output is the number on each dice and the total sum. 
    """
    dice_1 = []
    dice_2 = [] 
    sum_dice =[]
    for i in range(1,number+1):
        x = np.random.choice(range(1,7),p = prob)
        y = np.random.choice(range(1,7),p = prob)
        z = x+y
        dice_1.append(x)
        dice_2.append(y)
        sum_dice.append(z)


    result = pd.DataFrame({"Dice 1":dice_1,"Dice 2":dice_2,"Roll":sum_dice})
    result.index += 1
    return result
```

```python
def norm_dist (number):
    """
    Fair roll of 2 dices: Inputs are the number of rolls, 
    and the output is the number on each dice and the total sum. 
    """
    random = []
    density =[]
    for i in range(1,number+1):    
        x = np.random.uniform(4.5,6.7)
        y = (6.7-4.5)*norm(5.80,2.30).pdf(x)
        random.append(x)
        density.append(y)


    result = pd.DataFrame({"Rand":random,"Density":density})
    result.index += 1
    return result
```

```python
def test_for_norm(trial,exact,conf):
    """ The inputs for this function are the number of trials
    the exact value of the integral and the confidence interval
    expressed as a number of standard deviations from the mean.
    """
    fails = 0
    for i in range(1,trial+1):
        x = norm_dist(50)
        mean_x = x["Density"].mean()
        std_x = x["Density"].std()
        r1 = exact - (conf * std_x)
        r2 = exact + (conf * std_x)
        if mean_x <= r1:
            fails = fails + 1
        elif mean_x >= r2:
            fails = fails + 1 
    return fails
```

In this problem, we are analysing a simple day to day business of selling oats, peas, beans, and barley. For this project, I chose to use a For Loop in R. For me, this is an obvious choice, albeit a somewhat of a extravagent step. For the most part For loops are a thing of the past, and tend to gummy up the work. However, I feel that for this class especially, for-loops are an intuitive method for wrapping one's head around a simulation, as opposed to using the apply function or other ways to create just a mass block of "simulated data."  

```python
#Initiating Variables
prod = ["oats","peas","beans","barley"]
cost = [1.05,3.17,1.99,0.95]
price = [1.29,3.76,2.23,1.65]
lbs_o = range(0,11)
lbs_p = range(0,9)
lbs_bn = range(0,15)
lbs_ba = range(0,12)
```

```python
#Creating a Function

def sum_sale(days):
    """
    The input on this function is the number of days in the season,
    which is a representation of the number of "runs" The output is a
    dataframe
    """
    o = []
    p = []
    bn = []
    ba = []
    for i in range(1,days+1):
        w = np.random.choice(lbs_o)
        x = np.random.choice(lbs_p)
        y = np.random.choice(lbs_bn)
        z = np.random.choice(lbs_ba)
        o.append(w)
        p.append(y)
        bn.append(y)
        ba.append(z)
        
    result = pd.DataFrame({"Oats":o,"Peas":p,"Beans":bn,"Barley":ba})
    result.index += 1
    tot_cost = np.multiply(result, cost).sum(axis=1)
    tot_price = np.multiply(result, price).sum(axis=1)
    tot_rev = tot_price - tot_cost
    result['Cost'] = tot_cost
    result['Price'] = tot_price
    result['Revenue'] = tot_rev
    
    return result
        
    
```

```python
results = sum_sale(90)
```

```python
results.head(10)
```

As you can see from the above, we have created a function that produces the required day to day sales, which is exactly what we need to show how much our seller makes per day. 

However, since we want to run a simulation of a 1000 summers, it is necessary to cut out some functions and produce a code that shows just the summer total values to compare the range of values over the course of multiple summers. To do this we are going to cut down the code tou just total values

```python
def sum_sale_int(itr,days):
    """
    The input on this function is the number of days in the season,
    which is a representation of the number of "runs" The output is a
    dataframe
    """
    cost_data = []
    price_data = []
    rev_data = []
    for i in range(0,itr+1):
        o = []
        p = []
        bn = []
        ba = []
        for i in range(1,days+1):
            w = np.random.choice(lbs_o)
            x = np.random.choice(lbs_p)
            y = np.random.choice(lbs_bn)
            z = np.random.choice(lbs_ba)
            o.append(w)
            p.append(y)
            bn.append(y)
            ba.append(z)
        
        result = pd.DataFrame({"Oats":o,"Peas":p,"Beans":bn,"Barley":ba})
        result.index += 1
        tot_cost = np.multiply(result, cost).sum(axis=1)
        tot_price = np.multiply(result, price).sum(axis=1)
        tot_rev = tot_price - tot_cost
        cost_data.append(sum(tot_cost))
        price_data.append(sum(tot_price))
        rev_data.append(sum(tot_rev))
        
       
    data = pd.DataFrame({"Cost":cost_data,"Price":price_data,"Revenue":rev_data})
    result.index += 1    
    return data
```
